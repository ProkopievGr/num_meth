```python
def f(x):
    return np.cos(x)**2 - 0.5*np.cos(x) + 1/18
```

```python
def plot(f, a, b, xticks=0.1, y_lim=None, figsize_=(15,8)):
    x = np.linspace(a, b, num = 1000)
    plt.figure(figsize=figsize_)
    plt.plot(x, f(x))
    plt.xticks(np.arange(a, b, xticks))
    plt.plot(x, np.zeros_like(x), c='black')
    if y_lim is not None:
        plt.ylim(y_lim)
```

```python
def bisec(f, a, b, eps):
    dx = float(b-a)/2
    mid = a + dx
    
    while float(b-a) > eps:
        if np.sign(f(a)) != np.sign(f(mid)):
            b = mid
        else:
            a = mid
        dx /= 2
        mid = a + dx
    return mid
```

```python
f_root1 = bisec(f, 0, 1.3, 1e-10)
f_root1
```

```python
f_root2 = bisec(f, 1.3, 2, 1e-10)
f_root2
```

```python
np.allclose(f_true_root1, f_root1, rtol=1e-10)
```

```python
np.allclose(f_true_root2, f_root2, rtol=1e-10)
```

```python
coeffs = [1, 7.809249, 16.28542, -2.771356, -27.95304, -11.33921]
```

```python
p = np.poly1d(coeffs)
```

```python
plot(p, -4, 4, y_lim=(-3, 3), xticks=0.5, figsize_=(15, 5))
```

```python
def simple_iter(p, a, b, eps=1e-8):
    deriv = np.polyder(p)
    M = np.max(deriv(np.linspace(a, b, num=1e2)))
    m = np.min(deriv(np.linspace(a, b, num=1e2)))
    
    alpha = 2/(M+m)
    q = (M-m)/(M+m)
    i = 0
    
    x_0 = a
    x_root = x_0 - alpha*p(x_0)
    while np.abs(x_root - x_0) > np.abs((1-q)*eps/q):
        x_0 = x_root
        x_root = x_root - alpha*p(x_root)
        i += 1
        
    return (x_root, i)
```

```python
intervals = [(-3.7, -3.5), (-3.4, -3), (-2, -1.5), (-0.5, 0), (1, 1.5)]
intervals
```

```python
roots = []
for interv in intervals:
    root, iterations = simple_iter(p, *interv)
    print(root, iterations)
    roots.append(root)
```

```python
true_roots = sorted(np.roots(p))
```

```python
np.allclose(true_roots, roots, atol=1e-8)
```

```python
def newton(p, x_0, eps=1e-8):
    deriv = np.polyder(p)
    x_root = x_0 - p(x_0)/deriv(x_0)
    
    i = 0
    while np.abs(x_root - x_0) > eps:
        x_0 = x_root
        x_root = x_root - p(x_root)/deriv(x_root)
        i += 1
        
    return (x_root, i)
```

```python
roots_newton = []
for interv in intervals:
    root, iterations = newton(p, interv[0])
    print(root, iterations)
    roots_newton.append(root)
```

```python
np.allclose(true_roots, roots_newton)
```

```python
p2 = np.poly1d([1, -2.7, 0, 0, 3])
print(p2)
```

```python
plot(p2, -5, 5, xticks=0.5, y_lim=(-3, 3))
```

```python
comp_roots_newton = []
init_approx = [1, 3, 1j, -1j]

for approx in init_approx:
    root, iterations = newton(p2, approx, eps=1e-5)
    print(root, iterations)
    comp_roots_newton.append(root)
```

```python
np.allclose(sorted(np.roots(p2)), sorted(comp_roots_newton), atol=1e-5)
```
