```python
%matplotlib inline
from scipy.optimize import fsolve
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ===================== PROBLEM 2.1.10 ===============

def lattice_search(f, a, b, lattice = 1e-3, eps = 1e-3):
    X = [a + lattice * i for i in range(int((b - a)/lattice))]
    ans = []
    for x in X:
        ans.append(fsolve(f, x)[0])
    ret = []
    for an in ans:
        if all(np.abs(x - an) >= eps for x in ret) and a <= an and an <= b:
            ret.append(an)
    ret.sort()
    print('>> scipy fsolve: \n', ret)
    return ret

def f_2_1_20(x): # on [0, 2]
    return np.cos(x)**2 + (1/np.sqrt(2) - 1/4) * np.cos(x) - 1/(4*np.sqrt(2))

def g_2_1_20(x): # on [0, 2]
    return np.cos(x)**2 - 1/2*np.cos(x) + 1/16

def f_2_3_4(x):
    return (x-1)*np.sinh(x+1)-x
```

```python
def simple_bisec(f, a, b, eps = 1e-16, null_approx = 1e-16, offset = 1e-2, max_iter = 1e4, break_positive = True):
    if break_positive == True and (np.sign(f(a)) == np.sign(f(b))):
        return None
    xl = a
    xr = b
    c = xl
    i = 0
    while (np.abs(xl - xr) >= eps) and i < max_iter:
        i += 1
        c = (xl + xr) / 2.0
        if np.sign(f(xl)) == np.sign(f(c)):
            xl = c
        else:
            xr = c
    return c

def bisec(f, a, b, eps = 1e-16, lattice = 1e-3, null_approx = 1e-16, merge_ratio = 1e-8):
    X = [a + lattice * i for i in range(int((b - a)/lattice))]
    X.append(b)
    ans = []
    for i in range(len(X) - 1):
        ans.append(simple_bisec(f, X[i], X[i + 1], eps, null_approx))
    ans = set(ans)
    if None in ans:
        ans.remove(None)
    ans = list(ans)
    ret = []
    for an in ans:
        if all(np.abs(x - an) >= merge_ratio for x in ret) and a <= an and an <= b:
            ret.append(an)
    ret.sort()
    print('>> lattice bisection method: \n', ret)
    return ret

print('f:\n')
lattice_search(f_2_1_20, 0, 2)
bisec(f_2_1_20, 0, 2)
print('g:\n')
lattice_search(g_2_1_20, 0, 2)
print('>> just bisection:\n', simple_bisec(g_2_1_20, 0, 2, break_positive = False))

# bisection does not work properly for g as it is positive definite
```

```python
def plot(f, a, b):
    x = np.linspace(a, b, num = 1000)
    plt.figure(figsize=(15,8))
    plt.plot(x, f(x))
    plt.xticks(np.arange(0,2,0.1))
    plt.plot(x, np.zeros_like(x), c='black', linewidth = 0.5)

plot(f_2_1_20, 0, 2)
plot(g_2_1_20, 0, 2)
```

```python
# ===================== (NOT)PROBLEM 2.2 =====================
def sq(x):
    return x**2

def D(f, x, ieps = 1e+4, n = 1000):
    summ = 0
    for i in range(1, n):
        summ += (f(x + i/(n * ieps)) - f(x)) * (ieps * n) / i
    for i in range(1, n):
        summ += (f(x) - f(x - i/(n * ieps))) * (ieps * n) / i
    return summ / (2*(n - 1))

def deriv(f, m = 10):
    return lambda x : D(f, x, n = m)

print('>> bisection in [1.2, 1.4]: \n', simple_bisec(f_2_1_20, a=1.2, b=1.4))

def newton_method(f, x0, eps = 1e-16 , iter_cnt = 1000, silent = False):
    x = x0
    xs = x - f(x)/D(f, x)
    i = 0
    while np.abs(x - xs) >= eps and i < iter_cnt:
        i += 1
        x = xs
        xs = x - f(x)/D(f, x)
    if silent == False:
        print('>> newton_method: \n', xs)
    return xs, i

newton_method(f_2_1_20, 1.3) # 1e+6 iterations

# max of deriv. of f_2_1_20 on [1.3, 1.35] is approx -0.87339
# min is approx. -0.95595
#(c) wolfram

# ===================== PROBLEM 2.3.4 =====================

q1 = (0.08256)/(0.87339 + 0.95595)
q2 = 0.12697
print('\nq1: ', q1, '\nq2: ', q2, '\n')
a = -2/(0.87339 + 0.95595)

def ψ_2_1_20(x):
    return x - a*f_2_1_20(x)

plot(deriv(ψ_2_1_20), 0.5, 1.5)

def strange_2_1_20(x):
    return x + np.sin(f_2_1_20(x) * 1.085707269405)

def simple_iteration(ψ, x0, q, eps = 1e-16, max_iter = 1000, silent = False):
    x = x0
    xs = ψ(x)
    i = 0
    while np.abs(xs - x) >= eps * (1 - q)/q and i < max_iter:
        i += 1
        x = xs
        xs = ψ(x)
    if silent == False:
        print('>> simple_iteration method: \n', xs)
    return xs, i

simple_iteration(ψ_2_1_20, 1.31, q = q1)
print('>> strange_iteration method: \n', simple_iteration(strange_2_1_20, 1.31, q = q1, max_iter = 100, silent = True))

# maximum of modulus of derivative in strange case is 0.12697
```

```python
# =============== SOME ADDITIONAL STUFF ================

def p_der(poly):
    return [(i + 1) * poly[i + 1] for i in range(len(poly) - 1)]

def rp_ev(poly):
    return lambda x : sum([x**i * poly[i] for i in range(len(poly))])

def c_prod(c1, c2):
    return (c1[0]*c2[0] - c1[1]*c2[1], c1[0]*c2[1] + c1[1]*c2[0])

def c_s(c1, c2):
    return (c1[0] + c2[0], c1[1] + c2[1])

def c_rs(c, r):
    return (c[0] + r, c[1])

def c_rp(c, r):
    return (c[0]*r, c[1]*r)

def c_sum(arr):
    s = (0, 0)
    for a in arr:
        s = c_s(s, a)
    return s

def c_n(c1, c2):
    return (c1[0] - c2[0], c1[1] - c2[1])

def c_inv(c):
    return (c[0] / c_norm(c)**2, -c[1] / (c_norm(c)**2))

def c_norm(c):
    return np.sqrt(c[0]**2 + c[1]**2)

def c_pow(c, n):
    p = (1, 0)
    for i in range(n):
        p = c_prod(p, c)
    return p

def to_c(r):
    return((r, 0))

def c_x(c):
    return c[0]

def cx(c): return c[0]
def cx(c): return c[1]

def cp_der(poly):
    return [c_prod((i + 1, 0), poly[i + 1]) for i in range(len(poly) - 1)]

def cp_ev(poly):
    return lambda x : c_sum([c_prod(c_pow(x, i), poly[i]) for i in range(len(poly))])

def CD(p, c):
    return cp_ev(cp_der(p))(c)

def CD_CP(p, c):
    return c_prod(CD(p, c), c_inv(cp_ev(p)(c)))

def CP_CD(p, c):
    cd = CD(p, c)
    if cd == (0,0):
        cd = (1e-8, 0)
    return c_prod(c_inv(cd), cp_ev(p)(c))

def n_root(c, n):
    return (np.arccos(c[0]/c_norm(c)) / n, np.arcsin(c[1]/c_norm(c)) / n)

f = [(0,0),(0,0),(1,0)]

print(CP_CD(f, (5,0)))
c_inv((10,0))
```
